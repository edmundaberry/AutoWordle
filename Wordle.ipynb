# Imports

```python
# I/O
import numpy as np

# Wordle
from wordle import RealGame, NOT_PRESENT, WRONG_PLACE, RIGHT_PLACE

# Plotting
%matplotlib inline
```

# Make the game

```python
game = RealGame(n_letters=5, n_tries=6)
game.show()
print("START GAME!")
```

# Round 1: Ask

```python
game.blacklist(["AEROS"])
word = game.ask()
```

# Round 1: Tell

```python
result = np.array([NOT_PRESENT, NOT_PRESENT, WRONG_PLACE, NOT_PRESENT, NOT_PRESENT])
game.guess(word=word, result=result)
game.show()
game.evaluate()
```

# Round 2: Ask

```python
word = game.ask()
```

# Round 2: Tell

```python
result = np.array([NOT_PRESENT, RIGHT_PLACE, WRONG_PLACE, WRONG_PLACE, NOT_PRESENT])
game.guess(word=word, result=result)
game.show()
game.evaluate()
```

# What candidates are left?

```python
game.print_candidates()
```

# Some of these are very unlikely solutions

```python
game.blacklist(["HOING", "NOINT", "NOMOI", "POIND", "TOING"])
```

# Round 3: Ask

```python
word = game.ask()
```

# Round 3: Tell

```python
result = np.array([NOT_PRESENT, RIGHT_PLACE, RIGHT_PLACE, RIGHT_PLACE, NOT_PRESENT])
game.guess(word=word, result=result)
game.show()
game.evaluate()
```

# Last guess

```python
game.guess("POINT", result=[RIGHT_PLACE,] * 5)
game.show()
```
